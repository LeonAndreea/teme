## 3. Subtract the mean from each numerical column (standardize data)

Trageți media de la fiecare coloană numerică (standardizați datele)

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing
df3=df2.values
X_scaled = preprocessing.scale(df3)
X_scaled
```

```python
media=scaler.mean_ # extragem media pentru fiecare coloana 
print(media)
```

```python
standardizare_date=scaler.transform(df3)     # am standardizat datele 
print(standardizare_date)
```

## 4. Select top 5 numerical features that separate the groups in the "class" column

Selectați cele mai bune 5 caracteristici numerice care separă grupurile din coloana „clasă”

```python
colum_class=df.iloc[:, 81:82].values  # aceasta este coloana "class"
print(colum_class)
```

```python
## We are interested in the top 5 features that best discriminate between the classes
## This is a feature selection problem (as opposed to feature transformation accomplished through PCA)
## Since we didn't have time to cover this during the course, I will give you some hints

from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import SelectKBest

# use the mutual_info_classif criterion which measures the mutual information between a numerical feature and a categorical feature (our class column)
# SelectKBest will compute a discrepancy metric (in our case mutual_info_classif) between each numerical column and the class column and return the top k features with the hightest score

numFeatures = 5   # dorim sa extregem doar 5 caracteristici
targetData = colum_class
numericalData = standardizare_date  

best_features = SelectKBest(mutual_info_classif, k=numFeatures).fit_transform(numericalData,targetData)
print(best_features)  # se vor afisa cele 5 coloane 
```

```python
# putem pune cele 5 coloane intr-un DataFrame pentru a putea realiza scatter matrix 
import pandas as pd
df_best_features=pd.DataFrame(best_features)
df_best_features
```

## 5. Show the scatter matrix for the selected features (with hue = class column) 

Afișați matricea de împrăștiere pentru caracteristicile selectate (cu nuanță = coloana clasei)

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="ticks", color_codes=True)
sns.pairplot(df_best_features, hue='class')
plt.show()
```

## 6. Perform PCA on all standarized numerical features

Efectuați PCA pe toate caracteristicile numerice standardizate

```python
# pentru a efectua PCA trebuie sa standardizam datele ( ceea ce am facut-o mai sus), sa calculam matricea de covarianta,
# iar apoi calculam vectorii proprii
import numpy as np
from sklearn.decomposition import PCA

# 1. standardizarea datelor ( este realizata mai sus)
# 2. calcularea matricea de covarianta cu 5 caracteristici
covar_matrix = PCA(n_components = 5)
covar_matrix.fit(df_best_features)
componente=pd.DataFrame(covar_matrix.components_)
print(componente)
```

```python
# 3. calculam vectorii proprii 
variance = pd.DataFrame(covar_matrix.explained_variance_ratio_ )
print(variance) 

var=np.cumsum(np.round(covar_matrix.explained_variance_ratio_, decimals=5)*100)
print(var) #suma cumulativă a variației explicată cu 5 caracteristici

```

```python
plt.ylabel('Variance Explained')
plt.xlabel('Features')
plt.title('PCA ')
plt.plot(var)
```
